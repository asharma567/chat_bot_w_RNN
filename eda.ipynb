```python
'''
- fix the typos
    - autocomplete? spell check?
        https://github.com/mattalcock/blog/blob/master/2012/12/5/python-spell-checker.rst
        http://norvig.com/spell-correct.html
    - Remove Punctuation?
- removing infrequent terms?
- grammar
- removing names?
'''

all_oneway_customer_service_chats = []

for chat in chats:
    
    #for one way convos we just make a small adjust to to the filter
    customer_service_messages = [msg['Text'] for msg in chat if not msg['IsFromCustomer']]
    
    #preprocess and tokenize by sentence
    customer_service_messages_fmt = [sent_detector.tokenize(msg.strip().lower())
                                     for msg in customer_service_messages]
        
    all_oneway_customer_service_chats.append(customer_service_messages_fmt)

#unpack
line_by_line = [line for chat in all_oneway_customer_service_chats \
                for msg in chat for line in msg]    

#output
with open('oneway_customer_service_lines.txt','w') as f: 
    for line in line_by_line:
        f.write((line + '\n').encode('utf-8'))

!head -20 oneway_customer_service_lines.txt

with open('oneway_customer_service_lines2.txt','w') as f: 
    for chat in all_oneway_customer_service_chats:
        for msg in chat:
            for line in msg:
                f.write(line.encode('utf-8'))
                f.write('\n'.encode('utf-8'))
            f.write('\n\n')

!head -20 oneway_customer_service_lines2.txt
```

```python
all_twoway_customer_service_chats = []
for chat in chats:
    customer_service_messages = [message['Text'] for message in chat]
    all_twoway_customer_service_chats.append(customer_service_messages)
line_by_line = [msg for chat in all_twoway_customer_service_chats for msg in chat]    
with open('twoway_customer_service_lines.txt','w') as f: 
    for line in line_by_line:
        f.write((line + '\n').encode('utf-8'))
!head -20 twoway_customer_service_lines.txt

# output it file
with open('twoway_customer_service_lines2.txt','w') as f: 
    for chat in all_twoway_customer_service_chats:
        for msg in chat:
            f.write(msg.encode('utf-8'))
            f.write('\n'.encode('utf-8'))
        f.write('\n\n')

!head -20 twoway_customer_service_lines2.txt
```
